```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_game_info(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    title = soup.find('meta',attrs={'property':"og:title"}).get('content')
    tags = []
    for li in soup.find('ul', attrs={'class':"tags"}).find_all('li'):
        tags.append(li.text)
    rating = soup.find('span', attrs={'itemprop':"ratingValue"}).text
    rating_count = soup.find('span', attrs={'itemprop':"ratingCount"}).text
    return title, tags, rating, rating_count
```

```python
url = 'https://www.kongregate.com/games/Ninjakiwi/bloons-td-5'
```

```python
get_game_info(url)
```
