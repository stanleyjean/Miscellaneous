```python
import requests
from bs4 import BeautifulSoup
import konlpy.tag
import re
twitter= konlpy.tag.Twitter()
```

```python
def get_article(url):
    # Obtain three types of information about a news article
    headers = {
    'User-Agent': 'My User Agent 1.0',
    }
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'lxml')
    news_content = soup.find('div', attrs = {'id':'articleBodyContents'}).text
    news_content = news_content.split('{}')[1].strip()
    return news_content
```

```python
from ckonlpy.tag import Twitter
twitter = Twitter()
```

```python
twitter.add_dictionary('영탁', 'Noun')
```

```python
def kor_preprocessing(text, stopwords):
    filtered_content = re.sub('[^\s\w\d]',' ', content)
    Noun_words = twitter.nouns(filtered_content)
    unique_Noun_words = set(Noun_words)
    for word in unique_Noun_words:
        if word in stopwords:
            while word in Noun_words: Noun_words.remove(word)
    final_Noun_words = []
    for word in Noun_words:
        if len(word)>1:
            final_Noun_words.append(word)
    return final_Noun_words
```

```python
url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=101&oid=025&aid=0003102987'
content = get_article(url)
```

```python
print(content)
```

```python
stopwords = ['연합뉴스', '네이버', '기자']
final_nouns = kor_preprocessing(content, stopwords)
```

```python
from collections import Counter
c = Counter(final_nouns)
print(c.most_common(10))
```

## Word Cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from os import path

FONT_PATH = 'C:/Windows/Fonts/malgun.ttf' # For Korean characters

noun_text = ' '.join(final_nouns)
    
wordcloud = WordCloud(max_font_size=100, width=800, height=600, font_path=FONT_PATH)
wordcloud.generate(noun_text) 
# generate() 는 하나의 string value를 입력 받음
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
wordcloud.to_file('korean.png')
```
