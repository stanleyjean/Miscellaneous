```python
url = "http://www.yes24.com/Product/Goods/91065309"
```

```python
import requests 
```

```python
r = requests.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
soup
```

```python
soup.find('meta', attrs={"property":"og:title"}).get("content")
```

```python
soup.find("meta", attrs={"name":"author"}).get("content")
```

```python
soup.find("span", attrs={"class" : "gd_pub"}).text
```

```python
soup.find("span", attrs={"class":"gd_date"}).text
```

```python
soup.find("span", attrs={"class":"nor_price"}).text.rstrip("원")
```

```python
soup.find("span", attrs={"class":"nor_price"}).next_sibling.strip().strip("()")
```
