```python
url = 'https://www.imdb.com/title/tt0371746/'
```

```python
import requests
```

```python
r = requests.get(url)
```

```python
print(r.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

## 영화제목

```python
soup.find('title')
```

```python
soup.find('title').text
```

```python
title = soup.find('title').text
```

```python
title.split('-')
```

```python
title = title.split('-')[0].strip()
```

```python
title
```

## 평점

```python
soup.find('span', attrs={'itemprop':"ratingValue"})
```

```python
soup.find('span', attrs={'itemprop':"ratingValue"}).text
```

```python
rating = soup.find('span', attrs={'itemprop':"ratingValue"}).text
```

##  # of raters

```python
soup.find('span', attrs={'itemprop':"ratingCount"})
```

```python
num_raters = soup.find('span', attrs={'itemprop':"ratingCount"}).text
```

```python
num_raters
```

## # of reviewers

```python
soup.find('span', attrs={'itemprop':"reviewCount"})
```

```python
soup.find('span', attrs={'itemprop':"reviewCount"}).text
```

```python
num_reviewers = soup.find('span', attrs={'itemprop':"reviewCount"}).text
```

```python
num_reviewers = num_reviewers.split()[0]
```

```python
num_reviewers
```

## 영화감독 정보

```python
soup.find_all('div', attrs={'class':"credit_summary_item"})
```

```python
soup.find_all('div', attrs={'class':"credit_summary_item"})[0]
```

```python
soup.find_all('div', attrs={'class':"credit_summary_item"})[0].text
```

```python
soup.find_all('div', attrs={'class':"credit_summary_item"})[0].a
```

```python
soup.find_all('div', attrs={'class':"credit_summary_item"})[0].a.text
```

```python
director = soup.find_all('div', attrs={'class':"credit_summary_item"})[0].a.text
```

```python
content = ','.join([title, rating, num_raters, num_reviewers, director])
```

```python
content
```

```python
f = open('imdb_example.txt','w')
f.write(content)
f.close()
```
